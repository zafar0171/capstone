#### 5.Check if there are any non-real characters in the dataset.

```python
# Define a function to retrieve the non real characters from the dataframe
def check_non_real_characters(df):
    
    # List to store non-real characters
    non_real_characters = []

    # Iterate over each columns and each records
    for column in df.columns:
        for value in df[column]:
            
            # Check if any value is of string datatype and is not real
            if isinstance(value, str) and not np.isreal(value):
                # Apped those values which satisfies the above condition to the non-real characters list
                non_real_characters.append(value)

    # Return the list which contains the non-real characters            
    return non_real_characters

# Call the function which checks for any non-real characters present in the dataframe and returns it
non_real_chars = check_non_real_characters(df)

# Check if there are any non-real characters in the dataframe
if non_real_chars:
    print("Non-real characters found in the dataset:")
    print(non_real_chars)
else:
    print("No non-real characters found in the dataset.")
```

#### 10.Check the distribution of columns. List down columns that are normally distributed. List down columns that are not normally distributed.

```python
# Loop through the columns of the DataFrame
for column in df.columns:
    
    # Check if the column is numeric
    if df[column].dtype in [int, float]:
        
        # Calculate skewness and kurtosis
        skewness = df[column].skew()
        kurtosis = df[column].kurtosis()
        
        # Perform normality test (Shapiro-Wilk test)
        _, p_value = stats.shapiro(df[column])
        
        # Set a threshold for normality (e.g., p-value < 0.05)
        if p_value < 0.05:
            print(f"{column}: Not normally distributed (Skewness: {skewness}, Kurtosis: {kurtosis})")
        else:
            print(f"{column}: Normally distributed (Skewness: {skewness}, Kurtosis: {kurtosis})")
```

#### 11. List down columns that are highly skewed

```python
# Set the skewness threshold for identifying highly skewed columns
skew_threshold = 1.0  

# List to store highly skewed columns
highly_skewed_columns = []

# Loop through the columns of the DataFrame
for column in df.columns:
    
    # Check if the column is numeric
    if df[column].dtype in [int, float]:
        skewness = df[column].skew()
        
        # Check if the skewness is greater than the threshold
        if abs(skewness) > skew_threshold:
            highly_skewed_columns.append(column)

# Print the highly skewed columns
print("Highly skewed columns:")
for column in highly_skewed_columns:
    print(column)
```

#### 12. With the help of a standard scaler, normalize the respective column distribution.

```python
# Initialize standard scaler
scaler = StandardScaler()

# Loop through the highly skewed columns and apply the StandardScaler transformation:
for column in highly_skewed_columns:
    
    # Reshape the column values to a 2D array
    column_values = df[column].values.reshape(-1, 1)
    
    # Apply the StandardScaler transformation
    scaled_values = scaler.fit_transform(column_values)
    
    # Update the column values in the DataFrame
    df[column] = scaled_values
```

#### 17. Plot a heatmap for correlation matrix for the given dataset. Write the Observation. Especially note down columns that are highly correlated (Positive and Negative Correlation, Consider 0.7 to 1 as high).

```python
# Adjust plot size
plt.figure(figsize=(30, 20))

# Plot the heatmap with improved clarity
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Set the title and display the plot
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = df.corr()

# Set the threshold for high correlation
high_corr_threshold = 0.7

# Find the highly correlated columns
high_pos_corr = []
high_neg_corr = []

for col in corr_matrix.columns:
    high_pos_corr_cols = corr_matrix.index[corr_matrix[col] >= high_corr_threshold].tolist()
    high_neg_corr_cols = corr_matrix.index[corr_matrix[col] <= -high_corr_threshold].tolist()
    
    if col in high_pos_corr_cols:
        high_pos_corr_cols.remove(col)  # Remove the column itself from the list
    
    if col in high_neg_corr_cols:
        high_neg_corr_cols.remove(col)  # Remove the column itself from the list
    
    if high_pos_corr_cols:
        for col2 in high_pos_corr_cols:
            high_pos_corr.append((col, col2))
    
    if high_neg_corr_cols:
        for col2 in high_neg_corr_cols:
            high_neg_corr.append((col, col2))

# Print the highly correlated column pairs
print("High Positive Correlations:")
for col_pair in high_pos_corr:
    print(f"{col_pair[0]} and {col_pair[1]}")
    
print("\nHigh Negative Correlations:")
for col_pair in high_neg_corr:
    print(f"{col_pair[0]} and {col_pair[1]}")
```

#### 18.With the help of hypothesis testing check whether fraudulent transactions of higher value than normal transactions?

```python
# Separate the transaction values for fraudulent and normal transactions
fraudulent_values = df[df['Class'] == 1]['Amount']
normal_values = df[df['Class'] == 0]['Amount']

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(fraudulent_values, normal_values, equal_var=False)

# Adjust the p-value for a one-tailed test
if t_statistic > 0:
    p_value /= 2
else:
    p_value = 1 - p_value / 2

# Set the significance level (alpha)
alpha = 0.05

print('The p value is:',p_value)
# Check if the p-value is less than the significance level
if p_value < alpha:
    print("There is evidence to reject the null hypothesis.")
    print("Fraudulent transactions have higher values than normal transactions.")
else:
    print("There is no significant evidence to reject the null hypothesis.")
    print("Fraudulent transactions have lesser or equal values compared to normal transactions.")
```
